###  1. Introduccion

La estimación espectral consiste en calcular una aproximación del espectro de una señal a partir de un conjunto de muestras. Dado que se trabaja con 
señales finitas y discretas, toda estimación está sujeta a errores. El periodograma es la forma más simple de estimar la densidad espectral de potencia, obtenida al calcular el módulo al cuadrado de la transformada de Fourier de la señal. 

$$ P_{xx}(f) = |X(f)|^2$$

donde $ X(f) $ es la transformada de Fourier de $x(t)$. Sin embargo, esta técnica presenta una alta varianza, lo que puede dificultar la interpretación del espectro.

Para mejorar esta situación, se utilizan técnicas como el método de Welch, que consiste en dividir la señal en segmentos superpuestos, aplicar una ventana a cada segmento para reducir el efecto de discontinuidades, y luego promediar los periodogramas de cada bloque. Este enfoque reduce significativamente la varianza de la estimación, a costa de una leve pérdida de resolución espectral, lo cual representa un compromiso favorable en muchos casos prácticos.


Por otro lado, la función de autocorrelación de una señal real $ x(t) $, definida como:

$$ R_{xx}(\tau) = \int_{-\infty}^{\infty} x(t) \, x(t+\tau) \, dt $$

describe cómo se relaciona la señal consigo misma con un cierto desfase $tau$. En particular, el valor de la autocorrelación en $tau = 0 $representa la energía total de la señal:

$$ R_{xx}(0) = \int_{-\infty}^{\infty} |x(t)|^2 \, dt $$

Este valor coincide exactamente con el lado izquierdo del teorema de Parseval, que afirma que:

$$\int_{-\infty}^{\infty} |x(t)|^2 \, dt = \int_{-\infty}^{\infty} |X(f)|^2 \, df $$

Donde  $X(f)$ es la transformada de Fourier de $ x(t) $. Es decir, la energía de la señal en el tiempo es igual a la energía en el dominio de la frecuencia. Entonces, a través de esta identidad se establece una conexión directa entre: l valor de la autocorrelación en cero, el teorema de Parseval y densidad espectral de potencia (PSD). 

$$ P_{xx}(f) = \mathcal{F}\left\{ R_{xx}(\tau) \right\} $$

Comprobar que ambas energías coinciden (dentro de un margen aceptable de error) es un indicador de que la estimación espectral es consistente y que la implementación del método, incluyendo la normalización y la ventana, es adecuada.

### 2. Codigo

En este trabajo analizaremos cinco señales, cuyo comportamiento desconocemos a priori. Estas señales se clasifican en tres tipos: electrocardiograma, pletismografía y señales de audio. Para el análisis espectral, utilizaremos el estimador de Welch con una ventana de Hann.

La elección del estimador de Welch se debe a su baja varianza, lo que permite obtener una estimación más estable del espectro de potencia. La ventana de Hann contribuye a disminuir los efectos del solapamiento de los segmentos y a mejorar la resolución espectral frente a la fuga espectral. 

En este caso, cada segmento tiene una longitud de N/6 , lo cual representa un compromiso entre resolución y varianza. Además, se utiliza una FFT de longitud N (que corresponde a la longitud de la señal), y la segmentación se realiza a lo largo del eje temporal de la señal.

Importo librerias: 


```python
import numpy as np
from scipy import signal as sig
import matplotlib.pyplot as plt
import scipy.io as sio
```

Definicion de funciones útiles para todas las señales:

```python
def vertical_flaten(a):

    return a.reshape(a.shape[0],1)

def verificar_parseval(signal, nombre):
    area = np.sum(signal**2)
    #IMPORTANTE:Como mi señal se extrae con dimensiones [***, 1] mi fft no va a reconocer estas dimensiones porque trabaja por columnas, entonces va a interpretar que hay una sola columna! 
    #Conertimos el vector [N,1] a un vector de una sola dimension.
    ft = np.fft.fft(signal.flatten())
    parseval = np.mean(np.abs(ft)**2)
    error = np.abs(area - parseval) / area

    se_cumple = error < 0.01 
    
    return area, parseval, error, se_cumple

def calcular_anchodebanda(f, Pxx):
    # Paso 1: Potencia total acumulada
    potencia_total = np.sum(Pxx)
    
    # Paso 2: Potencia acumulada
    potencia_acumulada = np.cumsum(Pxx)
    
    # Paso 3: Umbrales 95% y 98%
    umbral_95 = 0.95 * potencia_total
    umbral_98 = 0.98 * potencia_total
    
    # Paso 4: Buscar índices donde se alcanza el umbral
    idx_95 = np.argmax(potencia_acumulada >= umbral_95)
    idx_98 = np.argmax(potencia_acumulada >= umbral_98)
    
    # Paso 5: Ancho de banda
    BW_95 = f[idx_95]
    BW_98 = f[idx_98]
    
    return BW_95, BW_98
```

### 2.1 ECG: Electrocardiograma

Una electrocardiografía es una forma de medir la actividad eléctrica del corazón, la cual se genera cuando las células cardiacas se contraen y relajan. El corazón puede simplificarse como un dipolo magnético, con un polo positivo y otro negativo. Cuando el corazón late, este dipolo cambia de dirección y magnitud. Su representación gráfica en el ECG se denomina trazado.



La onda P corresponde a la despolarización auricular, el complejo QRS representa la despolarización ventricular y ,finalmente, la onda T se vincula con la repolarización ventricular. En un ECG normal, estas ondas deben repetirse de manera rítmica y predecible. Cualquier alteración en la forma, duración o sincronía de estos componentes puede indicar problemas cardíacos como arritmias, isquemias, bloqueos de conducción, entre otros.

Qué se espera ver en un ECG:

* Señales con picos definidos (especialmente el complejo QRS).

* Onda P precediendo al QRS.

* Onda T posterior y de menor amplitud.

* Ritmo regular y constante (intervalos RR constantes).


```python
#Defino la frecuencia de muestreo:
fs_ecg = 1000 # Hz

#### ECG CON RUIDO ###

#Visualizamos lo que contiene el archivo
#io.whosmat('ECG_TP4.mat')

#Guardamos los datos del archivo en un diccionario de Python: enlistamos
mat_struct = sio.loadmat('./ECG_TP4.mat')

#Extraemos los datos de la señal y los metemos en un vector.
ecg_one_lead = vertical_flaten(mat_struct['ecg_lead'])
N1 = len(ecg_one_lead)

# Normalizo la señal 
ecg_one_lead_r = ecg_one_lead / np.std(ecg_one_lead)

### ECG SIN RUIDO ###
ecg_one_lead = np.load('ecg_sin_ruido.npy')
N2 = len(ecg_one_lead)
ecg_one_lead= ecg_one_lead / np.std(ecg_one_lead)
```

Estimacion por Welch

```python
f_ecg_r, Pxx_ecg_r = sig.welch(ecg_one_lead_r, fs_ecg, nfft=N1, window='hann', nperseg=N1//6, axis=0)
f_ecg, Pxx_ecg = sig.welch(ecg_one_lead, fs_ecg, nfft=N2, window='hann', nperseg=N2//6, axis=0)
print(np.max(Pxx_ecg_r))
print(np.max(Pxx_ecg))
```

Visualizacion de resultados:

```python
###SEÑAL TEMPORAL###
plt.figure(figsize=(10, 6))
#ECG con ruido
plt.subplot(2, 1, 1)  
plt.plot(ecg_one_lead_r)
plt.xlabel('tiempo [s]')
plt.ylabel('Amplitud [mV]')
plt.title("ECG: Electrocardiograma con ruido")

# ECG sin ruido
plt.subplot(2, 1, 2)
plt.plot(ecg_one_lead)
plt.xlabel('tiempo [s]')
plt.ylabel('Amplitud [mV]')
plt.title("ECG: Electrocardiograma sin ruido")

plt.tight_layout()
plt.show()

### ESPECTRO ###
#Grilla de frecuencias  
bfrec1= f_ecg_r <= fs_ecg / 2
bfrec2= f_ecg <= fs_ecg / 2

plt.figure(figsize=(10, 6))
#ECG con ruido
plt.subplot(2, 1, 1)  
plt.plot(f_ecg_r, 10 * np.log10(2 * np.abs(Pxx_ecg_r[bfrec1])**2))
plt.xlabel('frequency [Hz]')
plt.ylabel('PSD [V**2/Hz]')
plt.title('Power Spectral Density of: "ECG con ruido')

plt.subplot(2, 1, 2)  
plt.plot(f_ecg, 10 * np.log10(2 * np.abs(Pxx_ecg[bfrec2])**2))
plt.xlabel('frequency [Hz]')
plt.ylabel('PSD [V**2/Hz]')
plt.title('Power Spectral Density of: "ECG sin ruido')

plt.tight_layout()
plt.show()
```

Verifico Parseval

```python
area_r, parseval_r, error_r, cumple_r= verificar_parseval(ecg_one_lead_r, "ECG con ruido")
area, parseval, error, cumple = verificar_parseval(ecg_one_lead, "ECG sin ruido")
```

Busco Ancho de Banda

Asumimos comportamiento pasa bajos, asi tomamos el area acumulando la potencia con Cumsum. Es decir, toma el 95% del area bajo la curva desde 0 Hz

```python
BW_95_ecg_r, BW_98_ecg_r = calcular_anchodebanda(f_ecg_r, Pxx_ecg_r)
BW_95_ecg, BW_98_ecg = calcular_anchodebanda(f_ecg, Pxx_ecg)
```

Visualizacion de resultados

Tabla con los resultados ordenados

```python
fig, ax = plt.subplots(figsize=(10, 2))
ax.axis('off')
tabla = [
    ["Señal", "E Tiempo", "E Frecuencia", "Error", "¿Cumple Parseval?", "BW 95%", "BW 98%"],
    ["ECG con ruido", f"{area_r:.2f}", f"{parseval_r:.2f}", f"{error_r*100:.2f}%", "Sí" if cumple_r else "No", f"{BW_95_ecg_r:.1f}", f"{BW_98_ecg_r:.1f}" ],
    ["ECG sin ruido", f"{area:.2f}", f"{parseval:.2f}", f"{error*100:.2f}%", "Sí" if cumple else "No", f"{BW_95_ecg:.1f}", f"{BW_98_ecg:.1f}" ],
    ]
tabla_plot = ax.table(cellText=tabla, loc='center', cellLoc='center', colWidths=[0.14]*7)
tabla_plot.auto_set_font_size(False)
tabla_plot.set_fontsize(10)
tabla_plot.scale(1, 2)

plt.title("Verificación de Parseval y calculo de Ancho de Banda (por Welch)")
plt.show()
```

Grafico de la densidad espectral de potencia con el BW marcado

```python
plt.figure(figsize=(10, 6))
#ECG con ruido
plt.subplot(2, 1, 1)  
plt.plot(f_ecg_r, 10 * np.log10(2 * np.abs(Pxx_ecg_r[bfrec1])**2))
plt.axvline(BW_95_ecg_r, color='r', linestyle='--', label=f'BW 95%: {BW_95_ecg_r:.1f} Hz')
plt.axvline(BW_98_ecg_r, color='orange', linestyle='--', label=f'BW 98%: {BW_98_ecg_r:.1f} Hz')
plt.xlabel('frequency [Hz]')
plt.ylabel('PSD [V**2/Hz]')
plt.title('Power Spectral Density of "ECG" ')
plt.legend()
# ECG sin ruido 
plt.subplot(2, 1, 2)
plt.plot(f_ecg, 10 * np.log10(2 * np.abs(Pxx_ecg[bfrec2])**2))
plt.axvline(BW_95_ecg, color='r', linestyle='--', label=f'BW 95%: {BW_95_ecg:.1f} Hz')
plt.axvline(BW_98_ecg, color='orange', linestyle='--', label=f'BW 98%: {BW_98_ecg:.1f} Hz')
plt.xlabel('frequency [Hz]')
plt.ylabel('PSD [V**2/Hz]')
plt.title('Power Spectral Density of "ECG" ')
plt.legend()

plt.tight_layout()
plt.show()
```

### 2.2 PPG: Plestiograma

La PPG es una técnica de medición óptica de los cambios volumétricos de la sangre en áreas periféricas como pueden ser las yemas de los dedos o los lóbulos de las orejas. Estos cambios volumétricos son los que se producen debidos a los latidos del corazón, irrigando la sangre a todas las zonas de nuestro cuerpo.

El sistema está compuesto típicamente por dos diodos LED, uno de longitud de frecuencia correspondiente al color rojo y otro correspondiente al infrarrojo, que emiten luz pulsátil intermitente e intercaladamente. En el otro extremo del sensor hay un fotodetector que detecta aquella luz que no ha sido absorbida. De este modo, se puede componer una señal con componente DC y AC de la que puede extraerse información como el pulso cardíaco, la saturación de oxígeno en la hemoglobina y otros parámetros como la respiración. Un ciclo normal de una onda de ppg se ve de la siguiente manera: 



El trazado de una señal PPG típica incluye varios puntos clave:

* Pico sistólico: ocurre tras la contracción del ventrículo izquierdo, cuando la sangre es impulsada hacia las arterias.

* Valle dicroto: asociado al cierre de la válvula aórtica y al inicio de la relajación ventricular.

* Pico diastólico (a veces poco pronunciado): relacionado con el rebote de la onda de pulso en las paredes arteriales periféricas.

Algunos parámetros derivados del análisis de la señal incluyen:

* Ancho de pulso (PW50): es el tiempo que transcurre desde que la señal supera el 50 % del valor del pico sistólico en ascenso, hasta que lo vuelve a atravesar en descenso. Este valor se correlaciona con la resistencia vascular periférica.

* Intervalo de pulso: es el tiempo entre dos mínimos consecutivos en la señal de PPG. Este intervalo suele coincidir con el intervalo pico-a-pico, y es especialmente útil cuando los picos sistólicos no son fácilmente identificables debido a ruido o atenuación. 



```python
fs_ppg=1000 #Hz

# Extraigo los datos del archivo
ppg = np.load('ppg_sin_ruido.npy')
Np=len(ppg)

#Normalizo
ppg= ppg/np.max(ppg)
```

Estimacion por Welch

```python
f_ppg, Pxx_ppg = sig.welch(ppg, fs_ppg, nfft=Np, window='hann', nperseg=Np//40, axis=0)
print(np.max(Pxx_ppg))
```

Visualizacion de resultados:

```python
###Señal Temporal###
plt.figure()  
plt.plot(ppg)
plt.xlabel('tiempo [s]')
plt.ylabel('Amplitud [mV]')
plt.title("PPG: Pletismografía")

### Estrectro ###
bfrecp= f_ppg <= fs_ppg / 2

plt.figure()
plt.plot(f_ppg, 10 * np.log10(2 * np.abs(Pxx_ppg[bfrecp])**2))
plt.xlabel('frequency [Hz]')
plt.ylabel('PSD [V**2/Hz]')
plt.title('Power Spectral Density of "PPG" ')
```

Verifico Parseval y calculo Ancho de banda

```python
area_p, parseval_p, error_p, cumple_p = verificar_parseval(ppg, "PPG")
BW_95_ppg, BW_98_ppg = calcular_anchodebanda(f_ppg, Pxx_ppg)
```

Visualizacion de resultados:

```python
fig, ax = plt.subplots(figsize=(10, 2))
ax.axis('off')
tabla = [
    ["Señal", "E Tiempo", "E Frecuencia", "Error", "Cumple Parseval", "BW 95%", "BW 98%"],
    ["PPG sin ruido", f"{area_p:.2f}", f"{parseval_p:.2f}", f"{error_p*100:.2f}%", "Sí" if cumple_p else "No", f"{BW_95_ppg:.1f}", f"{BW_98_ppg:.1f}" ],
    ]
tabla_plot = ax.table(cellText=tabla, loc='center', cellLoc='center', colWidths=[0.14]*7)
tabla_plot.auto_set_font_size(False)
tabla_plot.set_fontsize(10)
tabla_plot.scale(1, 2)

plt.title("Verificación de Parseval y calculo de Ancho de Banda (por Welch)")
plt.show()
```

Grafico espectral marcando los ancho de banda

```python
plt.figure()
plt.plot(f_ppg, 10 * np.log10(2 * np.abs(Pxx_ppg)**2))
plt.axvline(BW_95_ppg, color='r', linestyle='--', label=f'BW 95%: {BW_95_ppg:.1f} Hz')
plt.axvline(BW_98_ppg, color='orange', linestyle='--', label=f'BW 98%: {BW_98_ppg:.1f} Hz')
plt.xlabel('frequency [Hz]')
plt.ylabel('PSD [V**2/Hz]')
plt.title('Power Spectral Density of "PPG"')
plt.legend()
plt.show()
```

### 2.3 Señales de audio

```python
fs_audio_c, wav_data_c = sio.wavfile.read('la cucaracha.wav')
fs_audio_s, wav_data_s = sio.wavfile.read('silbido.wav')
fs_audio_p, wav_data_p = sio.wavfile.read('prueba psd.wav')

N_c = len(wav_data_c)
N_s = len(wav_data_s)
N_p = len(wav_data_p)

# Normalizo en potencia: máximo entorno a -40dB
wavvv_data_c = wav_data_c / np.std(wav_data_c)
wavvv_data_s = wav_data_s / np.std(wav_data_s)
wavvv_data_p = wav_data_p / np.std(wav_data_p)

# Normalizo 2: Respecto al máximo, esto es para que todos tengan su pico al 0dB
# El ancho de banda nos juega en contra para comparar espectros de diferentes frecuencias
wavvv_data_c = wav_data_c / np.max(wav_data_c)
wavvv_data_s = wav_data_s / np.max(wav_data_s)
wavvv_data_p = wav_data_p / np.max(wav_data_p)
```

Estimación por Welch

```python
f_c, Pxx_c = sig.welch(wavvv_data_c, fs_audio_c, nfft=N_c, window='hann', nperseg=N_c//6, axis=0)
f_s, Pxx_s = sig.welch(wavvv_data_s, fs_audio_s, nfft=N_s, window='hann', nperseg=N_s//6, axis=0)
f_p, Pxx_p = sig.welch(wavvv_data_p, fs_audio_p, nfft=N_p, window='hann', nperseg=N_p//6, axis=0)
```

Visualizacion de resultados: 

```python
#### Señal Temporal ####
plt.figure(figsize=(10, 6))
plt.subplot(3, 1, 1)
plt.plot(wavvv_data_c)
plt.xlabel('tiempo [s]')
plt.ylabel('Amplitud [mV]')
plt.title("Audio.wav: La cucaracha")

plt.subplot(3, 1, 2)
plt.plot(wavvv_data_s)
plt.xlabel('tiempo [s]')
plt.ylabel('Amplitud [mV]')
plt.title("Audio.wav: Silbido")

plt.subplot(3, 1, 3)
plt.plot(wavvv_data_p)
plt.xlabel('tiempo [s]')
plt.ylabel('Amplitud [mV]')
plt.title("Audio.wav: Prueba")

plt.tight_layout()
plt.show()


##### ESPECTRO #######

plt.figure(figsize=(10, 6))
plt.subplot(3, 1, 1)
plt.plot(f_c, 10 * np.log10(2 * np.abs(Pxx_c)**2))
plt.xlabel('frequency [Hz]')
plt.ylabel('PSD [V**2/Hz]')
plt.title('Power Spectral Density of audio.wav of: "La cucaracha"')

plt.subplot(3, 1, 2)
plt.plot(f_s, 10 * np.log10(2 * np.abs(Pxx_s)**2))
plt.xlabel('frequency [Hz]')
plt.ylabel('PSD [V**2/Hz]')
plt.title('Power Spectral Density of audio.wav: "Silbido"')

plt.subplot(3, 1, 3)
plt.plot(f_p, 10 * np.log10(2 * np.abs(Pxx_p)**2))
plt.xlabel('frequency [Hz]')
plt.ylabel('PSD [V**2/Hz]')
plt.title('Power Spectral Density of audio.wav: "Prueba"')

plt.tight_layout()
plt.show()
```

Verifico Parseval

### 3. Análisis de Resultados

Algo que realmente me llamó la atención del análisis espectral de todas las señales fue cómo varía el comportamiento temporal cuando las señales poseen un alto número de muestras, lo que se traduce en un tiempo de observación más prolongado. Teniendo en cuenta la frecuencia de muestreo utilizada, si no se ajusta correctamente el eje temporal, la visualización puede ser errónea, ya que el tiempo de muestreo resulta demasiado corto para la cantidad total de muestras.
Por eso, es fundamental analizar el contenido de las muestras que componen cada señal, para ajustar adecuadamente los parámetros de visualización y así poder observar con mayor fidelidad la naturaleza real de la señal.

Siguiendo este mismo razonamiento, al realizar el análisis espectral mediante el método de Welch, la configuración de los bloques y el grado de solapamiento influyen directamente en el resultado espectral. Por ejemplo, si se utiliza una cantidad de bloques demasiado pequeña en relación con la duración total de la señal, se obtiene un espectro menos limpio, es decir, con irregularidades en la amplitud, debido a la pérdida de resolución espectral. Esto ocurre porque los bloques contienen demasiada información condensada, lo que impide una buena discriminación de frecuencias, generando un efecto de "ensanchamiento" en la representación espectral.

Por último, algo que claramente podría estar influyendo es el tema del filtrado de las señales. En este caso, no se aplicó ningún tipo de filtrado, y aunque algunas señales (como la del PPG) parecieran estar ya "filtradas", se desconoce completamente la naturaleza de ese filtrado
De hecho, se pudo observar que algunas señales aparecen atenuadas en amplitud, con frecuencias que no son típicas de su comportamiento fisiológico. Esto sugiere que el filtrado original podría haber sido ineficiente o mal ajustado. Además, se detectaron componentes de alta frecuencia en señales donde no deberían estar presentes, como en la PPG (fotopletismografía), que por su naturaleza debería tener contenido solo en bajas frecuencias.

Por todo esto, sería apropiado aplicar un filtro pasabajos, o al menos realizar un preprocesamiento básico que nos permita eliminar componentes espurios y ruido de alta frecuencia. De esa forma, podríamos trabajar exclusivamente con el rango de frecuencias que realmente nos interesa, mejorando tanto la claridad como la precisión del análisis espectral.

### 3.1 ECG

Analisis espectral y temporal: 
Se observa una diferencia clara en la amplitud: mientras que la señal sin ruido alcanza aproximadamente los 10 mV, la señal con ruido apenas llega a 5 mV. Esta atenuación puede deberse a una pérdida de información útil provocada por un filtrado excesivo o a que parte de la energía de la señal se encuentra enmascarada por el ruido presente.

Cuando se analiza el dominio temporal completo, estas diferencias se vuelven más evidentes. La señal sin ruido permite visualizar claramente los latidos cardíacos (con ondas R bien definidas), así como posibles latidos ectópicos. En contraste, la señal con ruido aparece completamente distorsionada y carece de una forma reconocible típica de un ECG.

Una razon que justificaria este comportamiento es la siguiente: a cantidad de muestras y duración total de las señales es muy diferente: la señal sin ruido contiene aproximadamente 30.000 muestras, lo cual equivale a 30 segundos si la frecuencia de muestreo es de 1000 Hz. Por otro lado, la señal con ruido contiene 1.129.116 muestras, lo que representa cerca de 19 minutos de grabación. Sin embargo, si se plotea esta última directamente sin ajustar el eje temporal ni limitar la cantidad de muestras, puede parecer que tiene sólo 1 segundo de duración, debido a la alta densidad de datos y al solapamiento visual.

Esta diferencia en duración y densidad de puntos explica por qué, al observar todo el eje temporal, las señales parecen tan distintas, mientras que en un rango acotado, su similitud es mucho mayor.

En resumen, cuando se grafican señales largas sin reducir la cantidad de puntos o sin ajustar el eje de tiempo, se pierde información visual importante. Por eso, no se puede graficar correctamente con la misma frecuencia de muestreo sin aplicar técnicas como recorte temporal (ventaneo) o downsampling
Podemos observar esta diferencia aplicando una restricción temporal adecuada en los gráficos:

```python
plt.figure(figsize=(10, 6))
#ECG con ruido
plt.subplot(2, 1, 1)  
plt.plot(ecg_one_lead_r[0:5000])
plt.title("ECG: Electrocardiograma con ruido")

# ECG sin ruido
plt.subplot(2, 1, 2)
plt.plot(ecg_one_lead[0:5000])
plt.title("ECG: Electrocardiograma sin ruido")

plt.tight_layout()
plt.show()
```

En el análisis espectral mediante la Densidad Espectral de Potencia (PSD), se observan picos destacados en ambas señales. En la señal sin ruido, los picos más relevantes se encuentran cerca de 0 Hz y 150 Hz. El primero puede deberse a la componente continua (offset), mientras que el segundo podría estar relacionado con un armónico del ritmo cardíaco o con restos de interferencia eléctrica no completamente eliminada.

Por otro lado, la señal con ruido muestra picos marcados en 50 Hz, 150 Hz y múltiplos, lo cual es característico de interferencia eléctrica de la red de alimentación. Esta observación confirma que el filtrado aplicado a esta señal no fue eficaz en eliminar completamente ese tipo de ruido, lo cual podría haberse corregido con un filtro notch en 50 Hz.

Finalmente, se observa una diferencia significativa en el ancho de banda (BW) entre ambas señales. La señal sin ruido posee un mayor ancho de banda (por ejemplo, BW95 ≈ 23 Hz), lo cual se corresponde con la presencia de mayor cantidad de detalles fisiológicos como las ondas P, QRS y T. En cambio, la señal con ruido muestra un ancho de banda más reducido (BW95 ≈ 17 Hz), producto de su distorsión y atenuación, lo que indica pérdida de contenido útil en frecuencias altas.
