## 1. Introduccion

### 1.1 Transformada Z

La transformada z es el análogo discreto de la Transformada de Laplace, y puede considerarse una generalización de la Transformada Discreta de Fourier (DFT), ya que introduce una variable compleja de módulo (radio) variable. La Transformada de Laplace, utilizada en sistemas continuos, permite analizar ecuaciones diferenciales a través de una variable compleja $ s=σ+jω $. La  Transformada Z permite representar señales discretas usando una variable compleja diferente:
$$ z= re^{jw} $$

Donde Donde:
* r es el módulo (radio) de la variable.
* ω es la frecuencia angular.
* j es la unidad imaginaria.
  
En funcion de esto, dada una funcion x(n) podemos definir formalmente su Transformada Z:
$$ X(z) = \sum_{n=\infty}^{\infty} x(n) z^{-n} $$

La Transformada Z bilateral surge de aplicar una serie de potencias generalizadas a una señal discreta. La elección de la suma bilateral permite representar cualquier tipo de señal discreta, no solo las causales.
$$ X(z) = \sum_{n=-\infty}^{\infty} x(n) z^{-n} $$

Para que la Transformada Z exista, es necesario que la serie de la suma converja. Esta región de convergencia (ROC) es el conjunto de valores de 
z para los cuales la serie infinita tiene un valor finito. Un sistema es estable si su ROC incluye el círculo unitario $∣z∣=1$. Es decir, si al evaluar la transformada sobre el eje $ z= re^{jw} $ la transformada está definida y no diverge. 


### 1.2 Demora, funcion transferencia y respuesta en frecuencia.

Un retardo (o demora) significa que estamos desplazando una señal, generalmente, hacia el plazo izquierdo. Esto nos permite representar sistemas cuasales, que dependen del tiempo presente y de entradas pasadas, pero no de salidas. 

La Transformada Z de una señal desplazada $ x(n−k)$ , donde k∈Z, se representa de la forma:
$$ \mathcal{Z}\{x(n - k)\} = \sum_{n = -\infty}^{\infty} x(n - k) z^{-n} $$

Si aca hacemos un cambio de variable: 
$$ m = n - k $$ 
$$ n = m + k$$

$$ \sum_{n = -\infty}^{\infty} x(n - k) z^{-n}
=  \sum_{m = -\infty}^{\infty} x(m) z^{-(m + k)}
=  z^{-k} \sum_{m = -\infty}^{\infty} x(m) z^{-m}
=  z^{-k} X(z) $$

A esto ultimo, lo conocemos como propiedad de retardo (o demora). Esto es particularmente util para el analisis de sistemas linealmente independientes (LTI). Ya que permite expresar ecuaciones diferenciales complejas, en ecuaciones simples. Para ello, se busca llegar a la exprecio de la funcion transferencia: 
$$ H(z) = \frac{Y(z)}{X(z)} $$
Donde $Y(z)$ es la funcion de salida y $X(z)$ la funcion de entrada.

Una forma típica de esta función para sistemas racionales es:
$$H(z) = \frac{b_0 + b_1 z^{-1} + b_2 z^{-2} + \cdots + b_M z^{-M}}{1 + a_1 z^{-1} + a_2 z^{-2} + \cdots + a_N z^{-N}}$$

Donde:
* $b_i$: coeficientes del numerador que determinan los ceros del sistema.
* $a_i$: coeficientes del denominador que determinan los polos del sistema.
* $z^-k $: representa un retardo de k muestras. 

Los polos y los ceros se grafican en el plano complejo Z. Donde el eje x es la parte real de z $ $  y el eje y es la parte imaginaria de z $jw$. Para analizar un poco mas la estabilidad del sistema, observamos donde se encuentran los polos respecto al circulo unitario:
- Si todos los polos están dentro del círculo unitario ($|z| < 1$), el sistema es estable.
- Si uno o más polos están fuera del círculo unitario ($|z| > 1$), el sistema es inestable.
- Si alguno está sobre el círculo ($|z| = 1$), el sistema es marginalmente estable..


La respuesta en frecuencia de un sistema LTI discreto se obtiene evaluando la función de transferencia sobre el círculo unitario. Es decir, evaluando en $z=1.e^{jw}$. 

$$ H(e^{j\omega}) = H(z)\big|_{z = e^{j\omega}}$$

Esta evaluación permite conocer cómo el sistema modifica el modulo y la fase de cada componente frecuencial $\omega$ de la señal de entrada.

Así, la respuesta en módulo y fase se calculan como:

$$ 
|H(e^{j\omega})| \quad;\quad
\arg\left(H(e^{j\omega})\right)
$$

### 2.1 Transformada

Los paso que se van a realizar para obtener la funcion transferencia son los siguientes:
* Aplicamos la transformada Z
* Extraemos factor comun, para obtener el cociente $\frac{Y(z)}{X(z)}$
* Extraemos factor común del exponente mayor, para obtener un cociente de polinomios. 

$ a. \quad y(n) = x(n - 3) + x(n - 2) + x(n - 1) + x(n)  $ 

1. Tranformo $$ \mathcal{Z}\{y(n)\} $$
$$ Y(z) = X(n)*z^{-3} + X(n)*z^{-2} + X(n)*z^{-1} + X(n) $$
3.  Funcion transferencia: $$ H(z) = \frac{Y(z)}{X(z)} = z^{-3} + z^{-2} + z^{-1} + 1 $$
4.  Mejoramos la expresion:
$$ H(z) 
= \frac{Y(z)}{X(z)}
=\frac{z^{3} + z^{2} + z^{1} + 1}{z^{3}} $$

$ b. \quad y(n) =  x(n - 4)  + x(n - 3) + x(n - 2) + x(n - 1) + x(n)  $ 

1. Tranformo $$ \mathcal{Z}\{y(n)\} $$
$$ Y(z) = X(n)*z^{-4} + X(n)*z^{-3} + X(n)*z^{-2} + X(n)*z^{-1} + X(n) $$
3.  Funcion transferencia: $$ H(z) = \frac{Y(z)}{X(z)} = z^{-4} + z^{-3} + z^{-2} + z^{-1} + 1 $$
4.  Mejoramos la expresion:
$$ H(z) 
= \frac{Y(z)}{X(z)}
=\frac{z^{4} + z^{3} + z^{2} + z^{1} + 1}{z^{4}} $$


$ c. \quad y(n) = x(n) - x(n - 1) $
 
1. Tranformo $$ \mathcal{Z}\{y(n)\} $$
$$ Y(z) = -X(n)*z^{-1} + X(n) $$
3.  Funcion transferencia: $$ H(z) = \frac{Y(z)}{X(z)} = -z^{-1} + 1 $$
4.  Mejoramos la expresion:
$$ H(z) 
= \frac{Y(z)}{X(z)}
= z^{-1} (z - 1) $$ 

$ d. \quad y(n) = x(n) - x(n - 2) $

1. Tranformo $$ \mathcal{Z}\{y(n)\} $$
$$ Y(z) = -X(n)*z^{-2} + X(n) $$
3.  Funcion transferencia: $$ H(z) = \frac{Y(z)}{X(z)} = -z^{-2} + 1 $$
4.  Mejoramos la expresion:
$$ H(z) 
= \frac{Y(z)}{X(z)}
= z^{-2} (z^2 - 1 ) $$ 

### 2.2 Respuesta de modulo y frecuencia

Los pasos para obtener la respuesta en frecuencia de modulo y fase son los siguientes: 
* Evaluar la función transferencia en $z=e^{jw}$
* Sacar factor común $e^{-j\omega \frac{N}{2}}$, donde $N$ es el grado máximo de retardo. Esto se hace especialmente en filtros no recursivos (FIR), ya que suelen tener simetría alrededor de la mitad de las muestras. Esto facilita la obtención del módulo y fase de la respuesta en frecuencia.
* Aplicamos la identidad de Euler segun sea correspondiente:
$$ \cos(\omega) = \frac{e^{j\omega} + e^{-j\omega}}{2} $$
$$ \sin(\omega) = \frac{e^{j\omega} - e^{-j\omega}}{2j} $$

$$ a. \quad H(z) = \frac{Y(z)}{X(z)} = z^{-1} (z - 1)$$
1.
$$ H(e^{j\omega})= e^{-j3\omega} (1 + e^{j\omega} + e^{j2\omega} + e^{j3\omega}) $$
2.
$$ H(e^{j\omega}) = e^{-j\frac{3\omega}{2}}(e^{-j\frac{3\omega}{2}} + e^{-j\frac{\omega}{2}} + e^{j\frac{\omega}{2}} + e^{j\frac{3\omega}{2}}) $$
3. 
$$ H(e^{j\omega}) = e^{-j\frac{3\omega}{2}}[2\cos(\frac{3\omega}{2}) + 2\cos(\frac{\omega}{2})]$$

Donde:

* $$|H(e^{j\omega})| = 2\cos( \frac{\omega}{2}) + 2\cos(\frac{3\omega}{2}) $$
* $$ \angle H(e^{j\omega}) = -\frac{3\omega}{2} $$ 



$$ b. \quad H(z)  = \frac{Y(z)}{X(z)} =\frac{z^{4} + z^{3} + z^{2} + z^{1} + 1}{z^{4}} $$

1. 
$$ H(e^{j\omega}) = e^{-j4\omega}(1 + e^{j\omega} + e^{j2\omega} + e^{j3\omega} + e^{j4\omega}) $$
2.
$$ H(e^{j\omega}) = e^{-j2\omega}\left(e^{-j2\omega} + e^{-j\omega} + 1 + e^{j\omega} + e^{j2\omega}\right) $$
3.
$$ H(e^{j\omega}) = e^{-j2\omega}[2\cos(2\omega) + 2\cos(\omega) + 1]$$

Donde: 
* $$ |H(e^{j\omega})| = 2\cos(\omega) + 2\cos(2\omega) + 1 $$
* $$ \angle H(e^{j\omega}) = -2\omega $$

$$ c. \quad H(z) 
= \frac{Y(z)}{X(z)}
= z^{-1} (z - 1) $$ 

1. $$ H(e^{j\omega}) = 1 - e^{-j\omega}$$ 
2. $$  H(e^{j\omega})= e^{-j\frac{\omega}{2}}(e^{j\frac{\omega}{2}} - e^{-j\frac{\omega}{2}}) $$
3. $$  H(e^{j\omega})= 2j\,e^{-j\frac{\omega}{2}} \sin(\frac{\omega}{2})$$

Donde: 
* $$ |H(e^{j\omega})| = 2\left|\sin\left( \frac{\omega}{2} \right)\right| $$
* $$ \angle H(e^{j\omega}) = -\frac{\omega}{2} + \frac{\pi}{2} $$ 

$$ d. \quad H(z) 
= \frac{Y(z)}{X(z)}
= z^{-2} (z^2 - 1 ) $$

1. $$ H(e^{j\omega}) = 1 - e^{-j2\omega} $$
2. $$  H(e^{j\omega})=e^{-j\omega}(e^{j\omega} - e^{-j\omega}) $$
3. $$  H(e^{j\omega})=2j\,e^{-j\omega} \sin(\omega)$$

Donde: 
* $$|H(e^{j\omega})| = 2|\sin(\omega)|$$
* $$\angle H(e^{j\omega}) = -\omega + \frac{\pi}{2}$$


### 2.3 Simulación 

Importo librerias:

```python
import numpy as np
from scipy import signal as sig
import matplotlib.pyplot as plt
```

Datos a simular:

```python
# Frecuencia
omega = np.linspace(0, np.pi, 500)

# Funciones transferencia (respuesta en frecuencia)
sistemas = {
    'H_a': 1 + np.exp(-1j*omega) + np.exp(-2j*omega) + np.exp(-3j*omega),
    'H_b': 1 + np.exp(-1j*omega) + np.exp(-2j*omega) + np.exp(-3j*omega) + np.exp(-4j*omega),
    'H_c': 1 - np.exp(-1j*omega),
    'H_d': 1 - np.exp(-2j*omega)
}
# Coeficientes 

numeradores = {
    'H_a': [1, 1, 1, 1],           
    'H_b': [1, 1, 1, 1, 1],             
    'H_c': [1, -1], 
    'H_d': [1, 0, -1]
}
denominadores = {
    'H_a': [1, 0, 0, 0],
    'H_b': [1, 0, 0, 0, 0],
    'H_c': [1, 0],
    'H_d': [1, 0, 0]
}
```

Visualizacion de resultados:

```python
for key in sistemas:
    H = sistemas[key]
    H_mod = np.abs(H)
    H_phase = np.angle(H)
    
    num = numeradores[key]
    den = denominadores[key]
    zeros, poles, gain = sig.tf2zpk(num, den)
    print(zeros)
    
    # Crear figura con 3 subgráficos
    fig, axs = plt.subplots(1, 3, figsize=(15, 4))
    
    # Módulo
    axs[0].plot(omega, H_mod)
    axs[0].set_title(f'Módulo |H(e^{{jω}})| - {key}')
    axs[0].set_xlabel('ω (radianes)')
    axs[0].set_ylabel('Módulo')
    axs[0].grid(True)
    
    # Fase
    axs[1].plot(omega, H_phase)
    axs[1].set_title(f'Fase ∠H(e^{{jω}}) - {key}')
    axs[1].set_xlabel('ω (radianes)')
    axs[1].set_ylabel('Fase (rad)')
    axs[1].set_xlim([0, np.pi])
    axs[1].grid(True)
    
    # Polos y ceros
    axs[2].scatter(np.real(zeros), np.imag(zeros), marker='o', s=80, label='Ceros', color='blue')
    axs[2].scatter(np.real(poles), np.imag(poles), marker='x', s=80, label='Polos', color='red')
    axs[2].add_artist(plt.Circle((0, 0), 1, fill=False, linestyle='--', color='gray'))
    axs[2].set_title(f'Plano-Z - {key}')
    axs[2].set_xlabel('Re')
    axs[2].set_ylabel('Im')
    axs[2].set_aspect('equal')
    axs[2].legend()
    axs[2].grid(True)
    axs[2].set_xlim([-1.5, 1.5])
    axs[2].set_ylim([-1.5, 1.5])
    
    plt.tight_layout()
    plt.show()
```
