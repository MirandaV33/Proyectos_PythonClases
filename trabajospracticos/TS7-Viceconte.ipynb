# 1. Introducción 

Un sistema LTI es un sistema que cumple con dos propiedades fundamentales: 

* Linealidad
* Invarianza en el tiempo

Estas dos propiedades implican que: responde linealmente a una combinación de señales y su comportamiento no cambia con el tiempo.

Una de las consecuencias más importantes de que un sistema sea LTI es que puede caracterizarse completamente por su respuesta al impulso, es decir, por cómo responde a una señal $\delta[n]$ que vale 1 en el instante $n = 0$ y 0 en todos los demás. Si se conoce esta respuesta, denominada $h[n]$, la salida del sistema para cualquier entrada $x[n]$ puede calcularse mediante la convolución:

$$ y(n) = x[n]*h[n] = \sum_{n=-\infty}^{\infty} x(n)h[n-k] $$. 

Este principio es la base del filtrado digital lineal, ya que el filtro puede verse como un sistema LTI. 

Para analizar estos filtros se utiliza la transformada Z, que permite estudiar el sistema en el dominio complejo. A través de ella se pueden identificar características como la respuesta en frecuencia, los polos y ceros, y la estabilidad. Además, en el dominio Z, las ecuaciones diferenciales se transforman en funciones racionales de la variable $z$, lo que simplifica mucho el análisis.

La ecuación general que describe un filtro digital es:


$$ y[n]  = \sum_{n=-\infty}^{\infty} b_i x[n-i]  - \sum_{n=-\infty}^{\infty} a_i y[n-i] $$

En esta ecuación, los coeficientes $b_i$ están asociados a las contribuciones de las entradas actuales y pasadas (parte no recursiva), mientras que los coeficientes $a_j$ corresponden a las salidas pasadas  Dependiendo de estos coeficientes, los filtros se pueden clasificar en:

1. Recursivos: que incluyen términos de retroalimentación a través de los coeficientes $a_j$. Estos pueden lograr respuestas más precisas usando menos coeficientes, pero su análisis es más complejo y pueden volverse inestables si no se diseñan correctamente. Su respuesta al impulso, en teoría, puede durar indefinidamente.
2. No recursivos: en los que $a_j = 0$ para todo $j$. Tienen entradas que no dependen de salidas anteriores. Por este motivo, siempre son estables y su diseño es relativamente sencillo.

Generalmente, los filtros no recursivos se conocen como Filtros de Respuesta al Impulso Finita (FIR), mientras que los filtros recursivos se denominan Filtros de Respuesta al Impulso Infinita (IIR). Esta clasificación proviene de cómo responden cuando se les aplica una señal impulso.


## Tipos de filtros: 

### IIR: Infinite Impulse Response

Su análisis no se basa tanto en la forma de h[n], sino en la posición de polos y ceros en el plano-Z. Pueden ser mucho más compactos (menos coeficientes para lograr la misma selectividad), pero no garantizan fase lineal ni estabilidad automática.
La función de transferencia de un filtro IIR en el dominio-Z es:
$$ H(z) = \frac{Y(z)}{X(z)} = \frac{\displaystyle\sum_{k=0}^{M} b_k z^{-k}}{\displaystyle\sum_{k=0}^{N} a_k z^{-k}} $$
 La posición de polos y ceros en el plano-Z determina las características del filtro. A diferencia de los FIR, los IIR pueden tener polos en ubicaciones distintas de cero, lo cual puede llevar a inestabilidad si los polos están fuera o sobre el círculo unitario. 
 No se puede garantizar fase lineal, lo que puede afectar señales sensibles a la distorsión de fase.


### Diseño de filtros IIR

El proceso generalmente implica diseñar un filtro analógico con las especificaciones deseadas (en el dominio de la frecuencia analógica) para luego transformar ese filtro al dominio digital.
Se utilizan 4 prototipos analogicos para el diseño de filtros IIR: 

* Butterworth: Proporciona una respuesta en la banda pasante lo más plana posible (maximally flat) y en la banda de atenuación, sin rizado (ondulaciones). Su principal desventaja es que requiere un orden mayor para lograr una transición abrupta entre bandas en comparación con otros tipos.

* Chebyshev Tipo I: Permite una transición más abrupta que Butterworth para un mismo orden, pero introduce rizado (ondulaciones) en la banda pasante.

* Chebyshev Tipo II (inverso): Es la contraparte del Tipo I, con rizado en la banda de atenuación pero una banda pasante plana.

* Elíptico: Ofrece la transición más abrupta posible para un orden dado, pero a cambio introduce rizado tanto en la banda pasante como en la banda de atenuación. Son los filtros de menor orden para una especificación dada.



Existen diferentes métodos para transformar el filtro analogico al dominio digital, el más usado es los siguientes: 

* Bilinear transform:
Este método permite convertir los polos del plano s (analógico) en polos dentro del círculo unitario en el plano z (digital). Lo logra sustituyendo:
$$ s = \frac{2}{T} \cdot \frac{z - 1}{z + 1} $$
En la funcion transferencia $H(s)$ para obtener $H(z)$. Esto mapea todo el semiplano izquierdo del plano s (donde residen los polos de los filtros analógicos estables) al interior del círculo unitario en el plano 'z' (donde deben residir los polos de los filtros digitales estables). Esto garantiza que un filtro analógico estable se transforme en un filtro digital estable.
Sin embargo, esta transformacion tiene una fuerte implicancia sobre la fase. El mapeado implicauna distorsion de frecuencias, ya que la conversion no es lineal. Esto afecta directamente dónde caen los puntos de corte, las bandas de paso y atenuación en la escala de frecuencia digital, es por esto que se debe de ajustar el diseño según las necesidades preferidas.

Estas transformaciones implican sustituciones específicas de la variable de frecuencia analógica s en la función de transferencia del prototipo. Luego, la transformada bilineal se aplicaría a la función resultante para obtener el filtro digital.

Por la naturaleza de los filtros IIR, podemos nombrar ciertas características de los mismos: 
* Fase no lineal: esto significa que diferentes componentes de frecuencia en la señal de entrada experimentan diferentes retardos de grupo, lo que puede distorsionar la forma de onda de la señal.
      ¿Que es un retardo de grupo?
      Implica que es la derivada negativa de la fase de la respuesta en frecuencia con respecto a la frecuencia no es constante para todas las frecuencias.
      $$\tau_g(\omega) = -\frac{d\theta(\omega)}{d\omega}$$
      Diferentes componentes de frecuencia de una señal de entrada se retrasan en diferentes cantidades de tiempo al pasar por el filtro. Esto distorsiona la forma de onda de la señal, incluso si la magnitud de la respuesta en frecuencia es ideal.
      La respuesta de fase no lineal en los IIR se debe a la estructura recursiva (retroalimentación) y la ubicación de sus polos y ceros (especialmente los polos) en el plano Z. Los filtros IIR no pueden tener fase lineal perfecta, excepto para el caso trivial de un solo polo o un solo cero.

  
  

  
  Por ejemplo, podemos observar que para un Butterworth, el retardo de grupo es más plano en la banda pasante que para los Chebyshev o Elípticos, lo que contribuye a su "suavidad" y mejor respuesta de fase (menos distorsión de forma de onda) en esa banda.

* Filtrado de fase 0: para aplicaciones no en tiempo real (donde se puede procesar una señal completa antes de que se necesite la salida), es posible lograr una respuesta de fase cero con un filtro IIR. Esto se hace aplicando el filtro en una dirección y luego, con la señal resultante, aplicando el mismo filtro en la dirección inversa. Al hacer esto, cualquier retardo de fase introducido por el primer paso se cancela por el segundo paso. La desventaja es que duplica el orden del filtro y no es causal. 


### FIR: Finitie Impulse Response

Los filtros FIR (Respuesta al Impulso Finita) que presentan fase lineal tienen una propiedad muy importante: sus coeficientes (la respuesta al impulso $h[n]$ exhiben simetría o antisimetría E.sta simetría, junto con si el número total de coeficientes N es par o impar, determina a qué tipo de filtro FIR pertenece

Existen cuatro tipos principales de filtros FIR con fase lineal:

* Simétrico, longitud impar:
La respuesta al impulso cumple: $h[n] = h[N - 1 - n]$
El número total de coeficientes N es impar.
* Simétrico, longitud par:
Se mantiene la simetría: $(h[n] = h[N - 1 - n]$ pero ahora N es par. Tienen un comportamiento diferente en frecuencia cerca de $\omega = \pi$, donde la magnitud puede anularse.
* Antisimétrico, longitud impar: 
En este caso:$ h[n] = -h[N - 1 - n]$ con N impar. Este tipo de simetría implica que $H(\omega)$ es **puramente imaginaria**.
* Antisimétrico, longitud par: 
También cumple: $ h[n] = -h[N - 1 - n]$ pero ahora N es par. En este caso, la magnitud se anula en \(\omega = \pi\), lo que limita su uso.

Resumiendo: 


| Tipo | Simetría       | Longitud \(N+1\) | β       | Fase lineal | $A(\omega)$                             | Coeficientes $a[m]$                |
|------|----------------|------------------|---------|-------------|--------------------------------------------|--------------------------------------|
| I    | Simétrica      | Impar            | 0       | Sí          | $ \sum_{m=0}^{M} a[m] \cos(m\omega)$      | $a[m] = h[M - m]$                |
| II   | Simétrica      | Par              | 0       | Sí          | $\sum_{m=1}^{M} b[m] \cos[(m - 1/2)\omega]$ | $b[m] = 2h[M - m]$                 |
| III  | Antisimétrica  | Impar            | $\pi/2$ | Sí        | $\sum_{m=1}^{M} c[m] \sin(m\omega)$      | $c[m] = 2h[M - m]$                 |
| IV   | Antisimétrica  | Par              | $\pi/2$ | Sí        | $\sum_{m=1}^{M} d[m] \sin[(m - 1/2)\omega]$ | $d[m] = 2h[M - m]$                 |


Cada combinación de simetría (par o impar) y longitud (par o impar) determina cómo se comporta el filtro en frecuencia, y qué tipo de filtro podemos construir con él: Los filtros simétricos tienden a tener fase lineal, lo cual es deseable porque no distorsionan la forma de la señal.

La simetría permite simplificar la Transformada Discreta de Tiempo (DTFT) del filtro: especialmente cuando se extrae el factor de fase:
    $$H(\omega) = e^{-j\omega M} \cdot H_r(\omega)$$
donde $M = \frac{N-1}{2}$ y $H_r(\omega)$ es la parte real de la respuesta, basada en la simetría de los coeficientes. Y cumple que:
    $$|H(\omega)| = |H_r(\omega)|$$
  Esto nos dice que toda la información sobre la magnitud de la respuesta en frecuencia está en la forma de los coeficientes, gracias a la simetría.

  

### Diseño de filtros FIR

El objetivo general del diseño de filtros FIR es encontrar los coeficientes $h[n]$ o $ b_k$ de un filtro de longitud finita tal que su respuesta en frecuencia $H(e^{j\omega})$, derivada de la Transformada de Fourier de Tiempo Discreto de $ h[n]$ , se aproxime lo mejor posible a una respuesta deseada  $H_d(e^{j\omega})$, manteniendo la propiedad de fase lineal.

* Ventanas 

Este es quizás el método más intuitivo y directo. Parte de la respuesta al impulso ideal de un filtro deseado (de duración infinita), que luego se "trunca" multiplicándola por una ventana de duración finita.

Ecuación general:

$$
h[n] = h_d[n] \cdot w[n]
$$

* Cuadrados Mínimos 

Este enfoque minimiza el error cuadrático medio entre la respuesta deseada y la real.

Función objetivo:

$$ E = \int_{\omega_p}^{\omega_s} \left| H_d(e^{j\omega}) - H(e^{j\omega}) \right|^2 \, d\omega $$



* Parks-McClellan

Este método busca minimizar el error máximo en bandas específicas de frecuencia, logrando ondulaciones iguales (equirepunte).

Problema de optimización:

$$ W(\omega) \cdot \left| H_d(e^{j\omega}) - H(e^{j\omega}) \right| $$


Estos cuatro métodos cubren la mayoría de las necesidades de diseño de filtros FIR. La elección depende de las especificaciones deseadas y la herramienta de diseño utilizada.


# 2. Código 

```python
#%% Librerias
import numpy as np
import scipy.signal as sig
import matplotlib.pyplot as plt
import scipy.io as sio

#Plantilla
from pytc2.sistemas_lineales import plot_plantilla
```

```python
#%% Definicion de funciones 
        
def plot_regions(ecg_signal, ecg_signal_filt, regs_interes, demora, label='ECG filtrado', crear_figura=True):
    
    cant_muestras = len(ecg_signal)
    for ii in regs_interes:
        zoom_region = np.arange(max(0, ii[0]), min(cant_muestras, ii[1]), dtype='uint')

        if crear_figura:
            plt.figure(figsize=(16, 8), facecolor='w', edgecolor='k')

        plt.plot(zoom_region, ecg_signal[zoom_region], label='ECG', linewidth=2)
        plt.plot(zoom_region, ecg_signal_filt[zoom_region + demora], label=label)
        
        plt.title('ECG filtering example from ' + str(ii[0]) + ' to ' + str(ii[1]) )
        plt.ylabel('Adimensional')
        plt.xlabel('Muestras (#)')
        
        axes_hdl = plt.gca()
        axes_hdl.legend()
        axes_hdl.set_yticks(())

        if crear_figura:
            plt.show()
```

```python
#%% Importamos los datos del ECG

#Guardamos los datos del archivo en un diccionario de Python: enlistamos
mat_struct = sio.loadmat('./ECG_TP4.mat')

#Extraemos los datos de la señal y los metemos en un vector.
ecg_one_lead = mat_struct['ecg_lead'].flatten()
N1 = len(ecg_one_lead)

# Normalizo la señal: 
ecg_one_lead = ecg_one_lead / np.std(ecg_one_lead)

# Defino variables que voy a analizar dentro del filtro.

# #Complejo de ondas QRS normal
qrs_pattern1 = mat_struct['qrs_pattern1'].flatten()
# Latido normal 
heartbeat_pattern1= mat_struct['heartbeat_pattern1'].flatten()
# Latido de origen ventricular
heartbeat_pattern2 = mat_struct['heartbeat_pattern2'].flatten()
# Vector con las localizaciones (en # de muestras) donde ocurren los latidos
qrs_detections = mat_struct['qrs_detections'].flatten()
```

Visualizacion de la señal en tiempo 

```python
plt.figure() 
plt.plot(ecg_one_lead, label= 'Señal de ECG completa')
plt.title("ECG: Electrocardiograma con ruido")
plt.xlabel('Tiempo')
plt.ylabel('Amplitud [V]')
plt.grid(True)
```

```python
seniales = {
    'QRS (latido normal)': qrs_pattern1,
    'Latido normal completo': heartbeat_pattern1,
    'Latido ventricular': heartbeat_pattern2,
    'Señal con detecciones QRS': qrs_detections  # para marcar qrs_detections
}

fig, axs = plt.subplots(2, 2, figsize=(12, 10), sharey=False)
axs = axs.flatten()  # para usar un solo índice: axs[0], axs[1], ...

for idx, (titulo, senial) in enumerate(seniales.items()):
    axs[idx].plot(senial, label=titulo)
    axs[idx].set_title(f'{titulo}')
    axs[idx].set_xlabel('Muestras')
    axs[idx].set_ylabel('Amplitud [V]')
    axs[idx].grid(True)
    axs[idx].legend()
plt.tight_layout()
plt.show()
```

Diseño de la plantilla: 

```python
#%% Diseño de la Plantilla del filtro --> Como queremos que se comporten nuestros filtros
# Para el ecg, tenemos ruido de baja frecuencia (ruido de linea y red electrica) y ruido de alta frecuencia. 
# Se diseña un filtro pasa banda. 

# Parámetros de la plantilla
fs = 1000 # Hz
nyq_frec = fs/2
ripple = 1 # dB
attenuation = 40 # dB

fpass = np.array([1.0, 35.0]) #Hz

#Atenuamos ruido de base de linea (0.1) y ruido de red electrica (50) junto con el de alta frecuencia
fstop = np.array([.1, 50.]) # Hz

plt.figure(figsize=(10, 6))
plt.title('Plantilla de Diseño de Filtro Pasabanda para ECG')
plt.xlabel('Frecuencia [Hz]')
plt.ylabel('Amplitud [dB]')
plot_plantilla(filter_type="bandpass", fpass=fpass, ripple=ripple, fstop=fstop, attenuation=attenuation, fs=fs)
plt.show()
```

### Diseño de filtros:

IIR: 

```python
#%% IIR 
aprox_name = ['butter', 'cheby1']
filtros_i= []

fig, axs = plt.subplots(1, 2,figsize=(12, 5), sharey=True)
axs = axs.flatten() 
# Eje de frecuencia
w_rad = np.append(np.logspace(-2, 0.8, 250), np.logspace(0.9, 1.6, 250))
w_rad = np.append(w_rad, np.linspace(40, nyq_frec, 500, endpoint=True)) / nyq_frec * np.pi

for idx, ftype in enumerate(aprox_name):
    mi_sos = sig.iirdesign(fpass, fstop, ripple, attenuation, ftype=ftype, output='sos', fs=fs)
    filtros_i.append(mi_sos)
    w, hh = sig.sosfreqz(mi_sos, worN=w_rad)
    axs[idx].plot(w / np.pi * fs / 2, 20 * np.log10(np.abs(hh) + 1e-15), label=ftype)
    axs[idx].set_title(f'Aproximación: {ftype}')
    axs[idx].set_xlabel('Frecuencia [Hz]')
    axs[idx].set_ylabel('Amplitud [dB]')
    axs[idx].grid(True)
    axs[idx].legend()
    plt.sca(axs[idx])
    plot_plantilla(filter_type="bandpass", fpass=fpass, ripple=ripple, fstop=fstop,
                   attenuation=attenuation, fs=fs)
plt.suptitle('Filtros IIR con diferentes métodos de aproximación', fontsize=16)
plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```

FIR: 

```python
freq = [0, fstop[0], fpass[0], fpass[1], fpass[1]+1, nyq_frec]
gain = [0, 0, 1, 1, 0, 0] 

# Ventanas
Filtro_Ventana = sig.firwin2(numtaps=2505, freq=freq, gain=gain, window='hamming', fs=fs)

# Cuadrados mínimos--> Requerimiento asimetric en frecuencia
# Opcion 1: Concatenar un pasa altos y un pasa bajos
# Opcion 2: cambiar la banda de paso! Hacerla simetrica 

numtaps_ls= 1505
Filtro_LS = sig.firls(numtaps_ls, freq, gain, fs=fs)

filtros_fir = {
    'Método de ventanas': Filtro_Ventana,
    'Método de cuadrados mínimos': Filtro_LS
}

# Visualización 
fig, axs = plt.subplots(1, 2, figsize=(12, 5), sharey=True)
axs = axs.flatten()

for idx, (nombre, filtro) in enumerate(filtros_fir.items()):
    w, h = sig.freqz(filtro, worN=8000, fs=fs)
    axs[idx].plot(w, 20 * np.log10(np.abs(h) + 1e-15), label=nombre)
    axs[idx].set_title(nombre)
    axs[idx].set_xlabel('Frecuencia [Hz]')
    axs[idx].set_ylabel('Amplitud [dB]')
    axs[idx].set_ylim(-100,10)  # Mostrará desde -100 dB hasta +30 dB
    axs[idx].grid(True)
    axs[idx].legend()
    # Plantilla en cada subplot
    plt.sca(axs[idx])
    plot_plantilla(filter_type="bandpass", fpass=fpass, ripple=ripple,
                   fstop=fstop, attenuation=attenuation, fs=fs)

plt.suptitle('Filtros FIR', fontsize=16)
plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```

## 3. Analisis de filtros: 

### IR: 

```python
ecgs_filtrados = {}  
nombres_filtros = ['Butter', 'Cheby I']

for nombre, sos in zip(nombres_filtros, filtros_i):
    ecg_filt = sig.sosfiltfilt(sos, ecg_one_lead)  # filtrar sin distorsión de fase
    ecgs_filtrados[nombre] = ecg_filt
    plt.figure() 
    plt.plot(ecg_one_lead, label= 'Señal sin filtrar')
    plt.plot(ecg_filt, label='Señal filtrada ')
    plt.title(f'ECG: Electrocardiograma filtrado por -{nombre}')
    plt.xlabel('Tiempo')
    plt.ylabel('Amplitud [V]')
    plt.grid(True)
    plt.legend()
    plt.show()
```

Analisis de regiones de interes: 

```python
regs_1 = (
    [10_000, 11_000],
    [300_000, 312_000],
    [720_000, 744_000]
)

demora = 0

for nombre, ecg_filt in ecgs_filtrados.items():
    fig, axs = plt.subplots(1, len(regs_1), figsize=(18, 5), sharey=True)
    axs = axs.flatten()
    
    for i, reg in enumerate(regs_1):
        plt.sca(axs[i])
        plot_regions(ecg_one_lead, ecg_filt, [reg], demora, label=nombre, crear_figura=False)
        axs[i].set_title(f'{nombre} - Región {i+1}')
        axs[i].set_xlabel('Muestras')
        axs[i].set_ylabel('Amplitud')
        axs[i].grid(True)

    plt.suptitle(f'{nombre} - Regiones de interes', fontsize=16)
    plt.tight_layout(rect=[0, 0, 1, 0.95])
    plt.show()


# Regiones con ruido
regs_2 = (
    [4000, 5500],
    [10000, 11000],
)
demora2=0


for nombre, ecg_filt in ecgs_filtrados.items():
    fig, axs = plt.subplots(1, len(regs_2), figsize=(12, 5), sharey=True)
    axs = axs.flatten()
    
    for i, reg in enumerate(regs_2):
        plt.sca(axs[i])
        plot_regions(ecg_one_lead, ecg_filt, [reg], demora2, label=nombre, crear_figura=False)
        axs[i].set_title(f'{nombre} - Región ruido {i+1}')
        axs[i].set_xlabel('Muestras')
        axs[i].set_ylabel('Amplitud')
        axs[i].grid(True)

    plt.suptitle(f'{nombre} - Regiones de interes', fontsize=16)
    plt.tight_layout(rect=[0, 0, 1, 0.95])
    plt.show()
    
```
